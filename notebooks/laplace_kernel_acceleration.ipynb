## Problem description

The problem at hand is very simple. Given a set of targets $X = \left\{\boldsymbol{x}_i \right\}_{i=1}^N$ and sources $Y = \left\{\boldsymbol{y}_i \right\}_{i=1}^N$ points, compute

\begin{align}
    \label{eq:main}
    u(\boldsymbol{x}) = \sum_{\boldsymbol{y} \in Y} G(\boldsymbol{x},\boldsymbol{y})\sigma(\boldsymbol{y})
    \quad \forall \quad \boldsymbol{x} \in X,
\end{align}

where $\sigma : Y \to \mathbb{R}$ is a known function, and
$$
    G(\boldsymbol{x},\boldsymbol{y}) = \frac{1} {\| \boldsymbol{x} - \boldsymbol{y} \|},
$$
is the (scaled) Greens function for Laplace's equation. 

The straigh-forward approach to solving this problem is to implement a double `for` loop: for every target point $\boldsymbol{x} \in X$, we compute a sum over $\boldsymbol{y} \in Y$. This algorithm has complexity $\mathcal{O}(MN)$. As we will see later on, depending on how far $Y$ is from $X$, we may exploit some mathematical properties of this problem in order to obtain a more efficient algorithm.

This notebook is split into two independent parts:
- [**Part I: Direct summation**](#direct-summation)
- [**Part II: Accelerated summation**](#accelerated-summation)

In the first part we consider the $\mathcal{O}(MN)$ algorithm, and explore various optimizations related to memory allocations, SIMD, and shared memory parallelism. In the second part we will look into an approximate way of performing such sums under suitable hypothesis on the points clouds $X$ and $Y$.

## [Part I: Direct summation](#direct-summation)

The first part is algorithmically simple: we will write a double loop to compute the potential $u(\boldsymbol{x})$ for every $\boldsymbol{x} \in X$. Sometimes there are better algorithms, as we will see in [Part II](#accelerated-summation), but sometimes you are stuck with the double `for` loop. Our goal here is to explore how to implement this **efficiently** in a way that exploits our hardware to the best of its ability.

We will consider 5 different versions of our algorithm:
1. A very high-level (and intentionally inneficient) implementation as you would probably write in `python` or `matlab`
2. A better implementation, similar (but more terse) to what you may write in `C` or `Fortran`
3. A threaded version to exploit shared memory parallelism
4. A vectorized version (in the sense of SIMD)
5. Finally, a threaded + vectorized version. 

Getting ahead of ourselves... at the end we should see something like the figure below, where our optimized code should run as much as $10x$ faster than a decent (but not vectorized/parallel) implementation!

### Naive implementation

In our first implementation, we will just write the algorithm as we wrote the formula, and not worry about things like memory allocation, cache locality, and parallelism. Our goal here is to get something that works, and therefore I will call this implementation `compute_interaction_naive!`. Here is how you could code it in `julia`:

```julia
"""
    compute_interaction_naive!(U,X,Y,σ)

Compute the interaction potential on locations `X::Matrix` 
as generated by sources `Y::Matrix` of intensity `σ::Vector`. 
The columns of the matrix denote the coordinates `x,y,z`, while 
the rows enumerate the particles.
"""
function compute_interaction_naive!(U,X,Y,σ)
    N = size(X,1)  # number of targets
    M = size(Y,1)  # number of sources
    for i in 1:N
        for j in 1:M
            U[i] += σ[j]/norm(X[i,:] - Y[j,:])
        end
    end
    return U
end
```

Note that I have chosen to represent the point clouds $X$ by a $N \times 3$ matrix `X`, and therefore the $k$-th coordinate of the $i$-th point is `X[i,k]`. Keep in mind that `julia` follows a [column major order](https://en.wikipedia.org/wiki/Row-_and_column-major_order), and therefore this means we store all the first coordinates of the points contiguously in memory, followed by the second coordinates, and so on. A valid alternative would be to represent this as a $3 \times N$ matrix instead, but as we will see later it is easier to vectorize the former representation. This is analogous to the [AoS vs SoA](https://en.wikipedia.org/wiki/AoS_and_SoA) choice. 

**Note**: in practice we should test our implementation to make sure it is *correct* before optimizing it. This algorithm is simple enough that I will just assume it is correct, and use it as a base to validate the other function we will code.

To talk about the performance we need to measure the execution of our programs. In `julia` there are several tools provided by the language to do that, and we will rely mosly on the macros `@time` and `@ellapsed`. You can bring up the help of a function in `julia` by typing `?` followed by the name of the function in a *REPL*, like done below:

```julia
??@elapsed
```

We will now create some random points and compute a vector of potential `U` using our `compute_interaction_naive!` function:

```julia
# create some random targets and sources
N,M = 1000,500
X = rand(N,3)
Y = rand(M,3)
σ = rand(M) # target densities
U = zeros(N) # vector to store output
# compute the interaction potential
compute_interaction_naive!(U,X,Y,σ); # to trigger compilation
@time compute_interaction_naive!(U,X,Y,σ);
```

On my computer this takes about a tenth of a second to finish, but the alarming sign here is the amount of (heap) allocations: about `1.5M allocations`. Remember that `julia` is a garbage collected language, so you don't have to explicitily handle the memory, but if you care about performance you should understand some of what is going on under the hood! 

In the `compute_interaction_naive!` function, it is easy to find the culprit: [*slicing* a vector](https://en.wikipedia.org/wiki/Array_slicing). The syntax `X[i,:]`, where `X` is a `Matrix`, will actually create a new `Vector` to store all the columns of the $i$-th row of `X`. The same would happen in `Matlab` by the way, so keep that in mind when you slice! 

The innermost part of our loop is allocating a bunch of small vectors, and this is bad news. Besides the overhead related to the dynamic allocation, this will also mean that the garbage collector will need to run often. In the next section we will see how to fix this.

### Fixing allocation issues

The *biggest problem* with our `compute_interaction_naive!` function is that it allocates *a lot* of memory.
The good news is that this is very easy to fix. The more *julian* way would probably be to use `view`s instead of the slicing (remember you can type `?view` on a `REPL` to see what it does), but we will take a `C`-like approach and just explicitly write the loops as follows:

```julia
"""
    compute_interaction!

An improved version of `compute_interaction_naive` which does not
allocate dynamic memory.
"""
function compute_interaction!(U,X,Y,σ)
    N,dim_X = size(X)
    M,dim_Y = size(Y)
    for i in 1:N
        for j in 1:M
            r2 = (X[i,1] - Y[j,1])^2 
            r2 += (X[i,2] - Y[j,2])^2
            r2 += (X[i,3] - Y[j,3])^2
            U[i] += σ[j]/sqrt(r2)
        end
    end
    return U
end
```

We can now test that this **computes the same result**:

```julia
U = zeros(N); V = zeros(N)
@test compute_interaction_naive!(U,X,Y,σ) == compute_interaction!(V,X,Y,σ)
```

Since we have no broken anything, let us time the execution again and compare it to the naive version:

```julia
t1 = @timed compute_interaction_naive!(U,X,Y,σ);
t2 = @timed compute_interaction!(V,X,Y,σ);
println("Runtime of Naive / regular =  $(t1.time/t2.time)")
println("Bytes allocated naive = $(t1.bytes)")
println("Bytes allocated regular = $(t2.bytes)")
```

Notice that this represents a huge improvement in the runtime, and it is mostly related to the memory allocation problem we had, so watch out for unnecesary allocations! We will worry about how the execution time scales relative to $M$ and $N$ later; as you may expect, the complexity should be $\mathcal{O}(N*M)$. The next optimization is relatively straigh-forward, and is related to shared memory parallelism (think [OpenMP](https://en.wikipedia.org/wiki/OpenMP)).

**Note**: You may be suspiscious that the problem is `julia`, and your code would go much faster in e.g. `C`. While the function `compute_interaction` should still be optmized a bit to be *in par* with a equivalent `C` implementation, its performance should not lag too much behind. The missing optimizations are related to [bounds checking](https://docs.julialang.org/en/v1/devdocs/boundscheck/), but for simplicity I have chosen to avoid getting into that. If you are interested/curious about implementations of this *direct summation* algorithm in various languages (and for a table of speed comparisons) see this nice [`repository`](https://github.com/ahbarnett/floatingspeed/tree/master/lap3dkernel) by Alex Barnett.  

### Multithreading

We will parallelize `compute_interaction!` by splitting the source points `X` into several groups, and creating a thread for each group which can be executed on a different processor. In `julia` you can use the `@threads` macro in front of a `for` loop to accomplish this. Think of a `parfor` loop in Matlab, or `#pragma omp for` if you are used to `OpenMP` in `C/C++`. This is probably the easiest extension, and here is the code:

```julia
function compute_interaction_threaded!(U,X,Y,σ)
    N,dim_X = size(X)
    M,dim_Y = size(Y)
    Threads.@threads for i in 1:N
        for j in 1:M
            r2 = (X[i,1] - Y[j,1])^2
            r2 += (X[i,2] - Y[j,2])^2
            r2 += (X[i,3] - Y[j,3])^2
            U[i] += σ[j]/sqrt(r2)
        end
    end
    return U
end
```

As before, lets make sure the answer is correct, and compare the performance:

```julia
U = zeros(N); V = zeros(N)
@test compute_interaction!(U,X,Y,σ) == compute_interaction_threaded!(V,X,Y,σ)
```

And do a rough comparison of performance:

```julia
t1 = @timed compute_interaction!(U,X,Y,σ);
t2 = @timed compute_interaction_threaded!(V,X,Y,σ);
println("Threaded version on $(Threads.nthreads()) threads")
println("Runtime single thread / multi thread =  $(t1.time/t2.time)")
println("Bytes allocated regular = $(t1.bytes)")
println("Bytes allocated threaded = $(t2.bytes)")
```

Therea are two important things to notice:
1. The threaded version is faster, but *not* 4 times faster. 
2. There is some memory allocation on the multithreaded version related to the handling of the threads.

Due to the overhead in multithreading, it is *not always faster* to run your code on multiple threads! Roughly speaking, you need to perform enough computations per thread to balance the overhead.

The next step is to exploit [**SIMD**](https://fr.wikipedia.org/wiki/Single_instruction_multiple_data) instructions; a.k.a. vectorization.

### Vectorization

There is one more important thing to talk about: vectorization. Your CPU has registers of various size, and ultimately data has to be moved to a registers before a computation can be performed on it. Some registers are big enough that you can fit a few double precision numbers on it. Some CPUs also have instructions which can operate on multiple data at the same time: [SIMD](https://fr.wikipedia.org/wiki/Single_instruction_multiple_data) for single instruction multiple data. This means if you do things right, you can e.g. substract four pairs of numbers at the same time! Our goal in is to exploit this that to effectively reduce the number of CPU instructions in our program.

Truth be told, this part is a bit harder than what we have considered so far. It requires a deeper understanding of the memory layout of your structures, but if you want optimized low-level code, you must think about how to vectorize your algorithm.

For our problem, the idea is simple: instead of looping over all $\boldsymbol{x} \in X$ and then summing over $\boldsymbol{y} \in Y$, we will do this in groups of `W` elements, where `W` is a integer representing the width of a "vector"; that is, how many primitive elements fit into one of my computer registers. For instance if I have a register of size `256` bits, so I can fit four double precision floats into it (`W=4`). 

The first thing we will consider is how to subtract four doubles with a single instruction. For this example, we will rely on `julia`'s automatic vectorization, which works quite well for simple examples. I will create two `Tuple` of size four, and write a function to subtract them. We will use the macro `@code_native` to print the compiled assembly code:

```julia
Xt = (1.,2.,3.,4.)
Yt = (3.,4.5,.3,8.)
vec_diff(X,Y) = X .- Y
@code_native debuginfo=:none vec_diff(Xt,Yt)
```

On my machine, the function `vec_diff` gets compiled to just a few assembly instructions, amongst which the most important one for our purposes is [`vsubpd`](https://software.intel.com/sites/landingpage/IntrinsicsGuide/#text=vsubpd). This instruction performs a *packed double-precision floating-point substraction*, and is exactly what we wanted our `julia` code to get compiled into. For simple operations on [`Tuple`](https://docs.julialang.org/en/v1/manual/functions/#Tuples) types, `julia` is quite good at automatically vectorizing code. Our actual application is a bit more involved than this simple example, and we will rely on the `julia` library `SIMD.jl` to make sure our code gets vectorized. This will also help us understand better what is going on, as we will have to write a few of the operations *by hand*. If you wanna try this in say `C`, I recommend looking for an appropriate SIMD library to avoid having to write intrinsics! 

To vectorize our function, we will break it down into a smaller components. First we will write a function which computes the potential 

```julia
using SIMD # a library for "manual" SIMD


"""
    _compute_pot_vec(Xt_vec::T,Yt_vec::T,Zt_vec::T,xs,ys,zs,q) where {T}

Compute the potential given a the coordinates of a vector of target points 
`Xt_vec,Yt_vec,Zt_vec`, and a source point `(xs,ys,zs)` of density `q`.
"""
function _compute_pot_vec(Xt_vec::T,Yt_vec::T,Zt_vec::T,xs,ys,zs,q) where {T}
    r2 = (xs - Xt_vec)*(xs - Xt_vec)
    r2 += (ys - Yt_vec)*(ys - Yt_vec)
    r2 += (zs - Zt_vec)*(zs - Zt_vec)
    pot_vec =  q / sqrt(r2)
    return pot_vec
end
```

To see how it works, we can call our function on a **vector** element as follows.

```julia
Xt = Yt = Zt = Vec(1.,2.,1.,1.)
xs = ys = zs = -1.0
q  = 1.0
_compute_pot_vec(Xt,Yt,Zt,xs,ys,zs,q)
```

More interestingly, we can quickly inspect the generated assembly code:

```julia
@code_native debuginfo=:none _compute_pot_vec(Xt,Yt,Zt,xs,ys,zs,q)
```

See all the `vmulpd`, `vaddpd`, `vsubpd`, `vsqrtpde`, etc? These are vectorized instructions we wanted! In summary our `_compute_pot_vec` takes the coordinates of 4 target points and a source point and compute a `<4xdouble>` vector representing the potential potential. The rest is just putting the pieces together. Here is the final vectorized code:

```julia
Base.@propagate_inbounds function _inner_loop(X1_vec,X2_vec,X3_vec,Y,σ)
    pot_vec = zero(X1_vec)
    @inbounds Y1 = view(Y,:,1)
    @inbounds Y2 = view(Y,:,2)
    @inbounds Y3 = view(Y,:,3)
    m = length(σ)
    for j in 1:m
        @inbounds pot_vec += _compute_pot_vec(X1_vec,X2_vec,X3_vec,Y1[j],Y2[j],Y3[j],σ[j])
    end
    return pot_vec
end
```

```julia
function compute_interaction_vectorized!(U,X,Y,σ)
    n = size(X,1)
    m = size(Y,1)
    V = Vec{4,Float64}
    X1 = view(X,:,1)
    X2 = view(X,:,2)
    X3 = view(X,:,3)
    for i in 1:length(V):n
        # load the data into `Vec` objects for vectorized computation
        X1_vec = vload(V,X1,i)
        X2_vec = vload(V,X2,i)
        X3_vec = vload(V,X3,i)
        pot_vec = _inner_loop(X1_vec,X2_vec,X3_vec,Y,σ)
        vstore(pot_vec,U,i)
    end
    return U
end
```

The only new things are:
1. We used`view` to create something akin to a pointer to each column of `X`.
2. Before looping over the source points, we loaded our data into a `Vec` type so as to perform vectorized operations on it.
3. After the loop, we used `vstore` to move the data back into `U`.

Yes, the code is more complex, but as we will see it is also significantly faster!

```julia
U = zeros(N); V = zeros(N)
# check for correctness
@test compute_interaction!(V,X,Y,σ) == compute_interaction_vectorized!(U,X,Y,σ)
U = zeros(N); V = zeros(N)
t1 = @timed compute_interaction!(V,X,Y,σ)
t2 = @timed compute_interaction_vectorized!(U,X,Y,σ)
println("Runtime regular / vectorized =  $(t1.time/t2.time)")
println("Bytes allocated regular = $(t1.bytes)")
println("Bytes allocated threaded = $(t2.bytes)")
```

On my machine, which supports `AVX` instructions on `256` bits registers, the speedup is about $2.5x$ times. Notice that unlike multithreading, vectorizing our code does not require managing threads and therefore does not allocate any dynamic memory. 

**Note:**
The term *vectorization* is sometimes used in a different context, specially in languages such as `matlab` or `python`, to mean that we will *avoid for loops* and put our data on *vectors* before performing arithmetic operations. This is associated with performance gain in interpreted languages because most linear algebra operations in these languages are actually calls to external libraries written in e.g. `C` or `FORTRAN`.

**Note:**
    One thing you have to be aware of when doing shared-memory parallelism is that everyone has access to the whole memory. This means you have to be careful when writing to memory to make sure you don't run into a *race condition*. In our example the value of $u$ at each target point $\boldsymbol{x}$ is modified by a single-thread, and therefore there is no issue whatsoever. If we  had chosen to parallelize the loop over the source points $\boldsymbol{y}$, we would have to be more careful. Race conditions lead to horrible bugs because the code works *sometimes*.

Finally, combining vectorization with the `@threads` macro is simple, so let us do it for completeness:

```julia
Base.@propagate_inbounds function compute_interaction_vectorized_threaded!(U,X,Y,σ)
    n = size(X,1)
    m = size(Y,1)
    V = Vec{4,Float64}
    X1 = view(X,:,1)
    X2 = view(X,:,2)
    X3 = view(X,:,3)
    Threads.@threads for i in 1:length(V):n
        # load the data into `Vec` objects for vectorized computation
        @inbounds X1_vec = vload(V,X1,i)
        @inbounds X2_vec = vload(V,X2,i)
        @inbounds X3_vec = vload(V,X3,i)
        pot_vec = _inner_loop(X1_vec,X2_vec,X3_vec,Y,σ)
        vstore(pot_vec,U,i)
    end
    return U
end

U = zeros(N); V = zeros(N)
@test compute_interaction!(V,X,Y,σ) == compute_interaction_vectorized_threaded!(U,X,Y,σ)
```

### Comparing different implementations for the direct summation part

We are now ready to compare everything we have coded so far. Remember we had 5 different versions:
    - `compute_interaction_naive!` : a first attempt at the problem, ignoring some simple performance issues like unnecessary memory allocations.
    - `compute_interaction!` : a rewrite of our first attempt in a more `C`-like manner.
    - `compute_interaction_threaded!` : partition the target points into different groups, and launch a thread on each.
    - `compute_interaction_vectorized!` : load the target and source coordinates four-at-a-time, and perform the operations using vectorized instructions.
    - `compute_interaction_vectorized_threaded!` : combine vectorization with shared memory parallelism.

To help visualize the differences, we will sweep the functions and the number of points from `M=N=100` to `M=N=2000` and collect the total time.

```julia
nrange   = 100:100:2000
T        = Dict()
flist = (compute_interaction_naive!,
        compute_interaction!,
        compute_interaction_threaded!,
        compute_interaction_vectorized!,
        compute_interaction_vectorized_threaded!)

for f in flist
    T[f] = []
    for n in nrange
        local X = rand(n,3); local Y = rand(n,3); 
        local σ = rand(n); local U = zeros(n)
        t = @elapsed f(U,X,Y,σ)
        push!(T[f],t)
    end
end

# you need to set the environment variable JULIA_NUM_THREADS=4 to run on 4 threads
Threads.nthreads() == 1 && @warn "You are running on a single thread"

fig1 = plot(xlabel="N",ylabel="seconds",legend=:topleft,ylim=(0,0.02))
fig2 = plot(xlabel="N",ylabel="normalized time",legend=:topleft,ylim=(0,30))
for f in flist
    plot!(fig1,nrange,T[f],m=:x,label="$f")
    plot!(fig2,nrange,T[f]./T[compute_interaction_vectorized_threaded!],m=:o,label="$f")
end
savefig(fig1,"runtime_comparison.png")
savefig(fig2,"relative_comparison.png")
```

That the naive implementation is off the charts should not come as a surprise given what we mentioned about memory allocation. Finally, before concluding this section, we mention that these are by no means all the optimizations we could have performed. If you want to dig deeper, here are some suggestions which may still give an extra performance boost:
- By default `julia` will perform [bounds check](https://docs.julialang.org/en/v1/devdocs/boundscheck/) when accesing an array. For performance critical parts of your code, you can disable that using the `@inbounds` macro. Alternatively you can launch `julia` with the option `--check-bounds=no` to disable **all** bounds check. Use with caution, as you could segfault and crash your program!
- Use `Float32` instead of `Float64` (i.e. `float` instead of `double` if you come from `C`). If you do this correctly, you should see a big improvement in your vectorized versions since you will be able to fit more numbers into a register.
- Enable `fastmath` (`-Ofast` in `gcc` jargon). If you allow the compiler to perform some operations which are true for real number but not for floating point numbers, you should be able to scratch a few percent speedup. But be careful: if your algorithm depends on e.g. the order in which operations are performed, `fastmath` will likely break it.

## [Part II: Accelerated summation](#accelerated-summation) 
<span>&#9888;</span> *Under construction*

While in [Part I](#direct-summation) we focused our efforts on decreasing the runtime by better exploiting the capabilities of our hardware, we will now work on the algorithmic complexity in order to reduce $N^2$ to something like $N \log(N)$ or even $N$. 

This is a different point of view as far as optimizing our solution goes. Instead of trying to make our algorithm run faster on a computer, we will try to make our algorithm better. The idea we will is a simplified version of what is behind the *fast multipole method*. 

### Low rank of far interactions

The first realization is that if the point clouds $X$ and $Y$ are sufficiently clustered and sufficiently, and their clusters are sufficiently far apart, then their interaction can in fact be well approximated by a low rank approximation. 

```julia
N = M = 1000
X = rand(N,3); Y = 10 .+ rand(M,3)
σ = rand(M)
scatter(X[:,1],X[:,2],X[:,3],label="targets",legend=:topleft)
scatter!(Y[:,1],Y[:,2],Y[:,3],m=:o,mc=:red,label="sources")
```

```julia
G(x,y) = 1/norm(x .- y)
A = [G(X[i,:],Y[j,:]) for i in 1:M, j in 1:N]
λ = eigvals(A;sortby=x->-abs(x))
scatter(norm.(λ),yscale=:log10,m=:x,label="eigenvalues")
```

```julia
function center_of_mass(X,σ)
    n  = length(σ)
    M  = 0.0
    xc = yc = zc = 0.0
    for i in 1:n
        xc += X[i,1]*σ[i]
        yc += X[i,2]*σ[i]
        zc += X[i,3]*σ[i] 
        M += σ[i]
    end
    return (xc/M,yc/M,zc/M),M
end

function center(X)
    n  = length(σ)
    xc = yc = zc = 0.0
    for i in 1:n
        xc += X[i,1]
        yc += X[i,2]
        zc += X[i,3]
    end
    return xc/n,yc/n,zc/n
end
```

```julia
n = m = 1000
X = rand(n,3)
σ = rand(m)
d = 2 .^ (1:10)
ee = []
for d in d
    Y = d .+ rand(m,3)
    U = zeros(n)
    compute_interaction!(U,X,Y,σ)
    Xc    = center(X)
    Yc,M  = center_of_mass(Y,σ)
    push!(ee,norm(U .- G(Xc,Yc)*M,Inf))
end
plot(d,ee,yscale=:log10,m=:x,xscale=:log10)
```
