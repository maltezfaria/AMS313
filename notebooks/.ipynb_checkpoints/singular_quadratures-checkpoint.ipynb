# Singular quadrature rules: change-of-variables

This notebook provides some supporting material for the first lecture of AMS313 on high-order boundary integral equation methods. The goal is to provide a hands-on experience to some of the techniques described, and verify how they work in practice. In particular we will focus on quadrature rules which are based on a change-of-variables the reference integration domain. That is, we will focus on integrals of the form:
\begin{align*}
\int_{\tau} K(x,y) \phi(y) ds(y)
\end{align*}
where $\tau$ is either a physical element, and the map $\chi : \hat{\tau} \to \tau$ is known. The cases were $\hat{\tau}$ is a reference line, reference triangle, or reference square, will be considered.

## Base cases

### Reference segment

#### Regular integrand

Let us verify the convergence of the Gauss-Legendre quadrature for a smooth function on $[0,1]$. For that we will consider
\begin{align}
\int_0^1 \cos(10x) dx = \sin(10)/10
\end{align}

```julia
k = 10
nn = 2:20
ee = []
f = s -> cos(k*s[1])
I = sin(k)/k
for n in nn
    quad = GaussLegendre(n)
    Iapprox = integrate(f,quad) # sum(f.(x).*w)
    er = abs((I - Iapprox)/I) + eps()
    push!(ee,er)
end
fig = plot(nn,ee,yscale=:log10,m=:x,label="Gauss-Legendre",legend=:bottomleft,xlabel="n",ylabel="error")
```

We can see that with about 13 points we achieve machine precision. To convince ourselves that not just any quadrature would achieve that, let us look at what the error would be if we used e.g. a trapezoidal rule:

```julia
ee = []
for n in nn
    quad = Trapezoidal(n)
    Iapprox = integrate(f,quad) #sum(f.(x).*w)
    er = abs((I - Iapprox)/I)
    push!(ee,er)
end
plot!(fig,nn,ee,yscale=:log10,m=:x,label="Trapezoidal")
```

For smooth functions, we see that Gauss quadrature converges exponentially fast (a straight line on a semi-log plot), while the trapezoidal rule should converge only as $1/n^2$. This shows the importance of chosing appropriate quadrature points and weights. There are other quadrature rules such as Fejer which achieves such a fast convergence, although Gauss is typically the best if all you care about is the the integration error:

```julia
ee = []
for n in nn
    quad = Fejer(n)
    x,w  = quad()
    Iapprox = sum(f.(x).*w)
    er = abs((I - Iapprox)/I)
    push!(ee,er)
end
plot!(fig,nn,ee,yscale=:log10,m=:x,label="Fejer")
```

```julia
xg,_ = GaussLegendre(20)()
xg = [xg[1] for xg in xg]
xx = 0:0.01:1
fig2 = plot(xx,f.(xx),lc=:black,label="")
scatter!(fig2,xg,f.(xg),m=:o,mc=:red,label="Gauss-Legendre")
xt,_ = Trapezoidal(20)()
xt = [xt[1] for xt in xt]
scatter!(fig2,xt,f.(xt),m=:x,mc=:blue,label="Trapezoidal")
```

#### Singular integrand

Let us see now what happens if we integrate instead 
\begin{align}
    \int_0^1 \log(x) dx = x \log(x) - x \Big|_{0}^1 = -1
\end{align}

```julia
nn = 2:40
ee = []
g  = s -> log(s)
I  = -1
for n in nn
    quad    = GaussLegendre(n)
    Iapprox = integrate(g,quad) # sum(g.(x).*w)
    er = abs((I - Iapprox)/I)
    push!(ee,er)
end
fig_sing = plot(nn,ee,yscale=:log10,m=:x,label="Gauss-Legendre",legend=:bottomleft,xlabel="n",ylabel="error")
```

This is much worse what what we observed for the smooth integrand. The lack of regularity of our kernel is the culprit, and the convergence is no better than linear.

```julia
order = [(nn[k+1]/nn[k]) /(ee[k+1]/ee[k])  for k in 1:length(ee)-1];
println("Estimated order for last point: $(order[end])")
```

We learned in the lecture that one way to recover a fast convergence is to modify our quadrature rule, designed for smooth functions, so that it **accumulates points near the singularity**. The proper way to accumulate points is through a degenerate change of variables $\phi$. We will consider two change of variables which we shall name:
- Kress
- IMT

<div class="alert alert-warning"> 
<b>Note: </b>
Feel free to consult the code to see the precise definition of these change of variables, as well as the parameters they contain (which have a value set by default in the constructors below)
</div>

```julia
kress = Kress()
imt   = IMT()
x = 0:0.01:1
fig = plot(x,kress.(x),label="Kress",lc=:red,legend=:topleft)
plot!(fig,x,imt.(x),label="IMT",lc=:blue)
```

The goal of such change of variables is to accumulate points near the endpoint singularity so as to improve the convergence of the quadrature rule. Note that an alternative would be locally refine our domain near the singularity. To see how this plays a role, let us consider what happends to e.g. the trapezoidal rule when the transformed under the `Kress` change-of-variables:

```julia
# plot nodes before and after change-of-variables
n = 10
x    = [i/n for i=0:n-1]
fig1 = plot(x,zero(x),m=:o,mc=:blue,label="Trapezoidal")
fig2 = plot(kress.(x),zero(x),m=:o,label="Trapezoidal+C.O.V",mc=:red)
plot(fig1,fig2)
```

These change of variables map $\hat{\tau} \to \hat{\tau}$, where $\hat{\tau} = [0,1]$, and have the following properties:

- Bijective
- Monotonically increasing
- Infinitely differentiable
- Degenerate at $s=0$

What is important to keep in mind is that they map integrals over $[0,1]$ into integrals over $[0,1]$, but they regularize values near $s=0$. 

We can now investigate how our modified quadrature performs:

```julia
shand = Kress()
ee = []
for n in nn
    quad_std = GaussLegendre(n)
    quad = SingularQuadratureRule(quad_std,shand)
    x,w  = quad()
    Iapprox = sum(g.(x).*w)
    er = abs((I - Iapprox)/I)
    push!(ee,er)
end
plot!(fig_sing,nn,ee,m=:x,label="Gauss-Legendre + Kress",legend=:bottomleft,ylim=(1e-15,0))
plot!([0,40],[5e-2,5e-2],ls=:dash,label="5%")
plot!([0,40],[1e-2,1e-2],ls=:dash,label="1%")
plot!([0,40],[1e-3,1e-3],ls=:dash,label="0.1%")
```

We see that, in this simple example, the conclusion is very different depending on *how accurate* we need our approximation to be. 

### Reference square

We now consider the case where $\hat{\tau}$ is a reference square.

#### Tensor-product rules

The easiest thing to do for the reference square is to simple tensorize one-dimensional quadrature rules. It is worth noting that, specially for low order quadratures, there are other choices which achieve a similar order of integration with fewer quadrature points. For high-order, however, the tensor-product technique is the usual choice.

```julia
q1d  = GaussLegendre(10)
quad = TensorProductQuadrature(q1d,q1d)
x    = qnodes(quad)
plot(Plots.Shape([0,1,1,0,0],[0,0,1,1,0]),label="",aspect_ratio=:equal)
scatter!(x,m=:x,mc=:black,label=nothing)
```

As before, we will demonstrate exponential convergence for smooth functions:

```julia
k = (10,5)
nn = 2:20
ee = []
f = s -> cos(k[1]*s[1])*cos(k[2]*s[2])
I = sin(k[1])/k[1]*sin(k[2])/k[2]
for n in nn
    q1d  = GaussLegendre(n)
    quad = TensorProductQuadrature(q1d,q1d)
    x,w  = quad()
    Iapprox = integrate(f,quad)
    er   = abs((I - Iapprox)/I) + eps()
    push!(ee,er)
end
fig = plot(nn,ee,yscale=:log10,m=:x,label="Gauss-Legendre",legend=:bottomleft)
```

#### Singular integrand

We consider now a singular integrand, where the singularity is located at the origin, and apply the standard quadrature rule to integrate it. As expected, we should see that the convergence order is largely reduced.

```julia
# construct a function singular at the origin
f = s -> begin
    d = sqrt(s[1]^2+s[2]^2)
    exp(10*im*d)/d
end
# integrate it using a tensor product rule and compute the error
I,E = WaveProp.Integration.hcubature(f,ReferenceSquare())
nn = 2:20
ee = []
for n in nn
    q1d  = GaussLegendre(n)
    quad = TensorProductQuadrature(q1d,q1d)
    x,w  = quad()
    Iapprox = integrate(f,quad)
    er   = abs((I - Iapprox)/I) + eps()
    push!(ee,er)
end
fig_sing = plot(nn,ee,xscale=:log10,yscale=:log10,m=:x,label="Gauss-Legendre",legend=:bottomleft,
                xlabel="n",ylabel="error")
```

Just like we tensorized the one-dimensional quadrature rules, we can tensorize the one-dimensional change-of-variables used to "regularize" the singular point. That is what we do in the next block, where we show that the ultimate effect of such change of variables is to modify the location of the nodes (and of course the value of the weights as well) so that we accumulate points near the singularity. 

```julia
# construct a singular quadrature
shand_1d = Kress(order=2)
shand    = TensorProductSingularityHandler(shand_1d,shand_1d)
q1d  = GaussLegendre(5)
qreg = TensorProductQuadrature(q1d,q1d)
quad = SingularQuadratureRule(qreg,shand)
# plot the nodes before and after the change of variables
x = qnodes(quad)
fig1 = plot(Plots.Shape([0,1,1,0,0],[0,0,1,1,0]),label="",aspect_ratio=:equal)
scatter!(qnodes(qreg),m=:o,mc=:black,label="",title="Gauss-Legendre")
fig2 = plot(Plots.Shape([0,1,1,0,0],[0,0,1,1,0]),label="",aspect_ratio=:equal)
scatter!(x,m=:o,mc=:red,label="",title="Gauss-Legendre + C.O.V.")
plot(fig1,fig2)
```

We can now verify how the modified quadrature rule performs compared to the *standard* quadrature:

```julia
ee = []
for n in nn
    q1d  = GaussLegendre(n)
    qreg = TensorProductQuadrature(q1d,q1d)
    quad = SingularQuadratureRule(qreg,shand)
    Iapprox = integrate(f,quad)
    er   = abs((I - Iapprox)/I) 
    push!(ee,er)
end
plot!(fig_sing,nn,ee,m=:x,label="Gauss-Legendre + Kress",legend=:bottomleft)
plot!([1,20],[5e-2,5e-2],ls=:dash,label="5%")
plot!([1,20],[1e-2,1e-2],ls=:dash,label="1%")
```

We see that both methods converge linearly (this is a log-log plot), but a much faster rate is observed for the corrected quadrature. 

### Reference triangle

Finally we consider the reference triangle, and look at a function which is singular at the $(1,0)$ vertex.

```julia
f = s -> begin
    d = sqrt((s[1]-1)^2+s[2]^2)
    exp(im*d)/d
end
I = integrate(f,ReferenceTriangle()) # reference value integrated using an adaptive method
```

We will use the Duffy change of variables. Here how the quadrature nodes look like after the transformation:

```julia
q1d   = GaussLegendre(5)
qstd  = TensorProductQuadrature(q1d,q1d)
duffy = Duffy()
quad  = SingularQuadratureRule(qstd,duffy)
x     = qnodes(quad)
plot(Plots.Shape([0,1,0,0],[0,0,1,0]),label="",aspect_ratio=:equal)
scatter!(x,m=:o,mc=:red,label="")
```

The convergence is exponential, as expected:

```julia
k = (10,5)
nn = 2:20
ee = []
shand    = Duffy()
ee = []
for n in nn
    q1d  = GaussLegendre(n)
    qreg = TensorProductQuadrature(q1d,q1d)
    quad = SingularQuadratureRule(qreg,shand)
    Iapprox = integrate(f,quad)
    er   = abs((I - Iapprox)/I) + eps()
    push!(ee,er)
end
plot(nn,ee,yscale=:log10,m=:x,label="Gauss-Legendre + Duffy",legend=:bottomleft)
```

## Reducing to base cases

In the [first section](#base-cases) we saw how to handle the base cases where the singularity lies on one a specific vertex of the reference element $\hat{\tau}}$. We will now show how to use this base case to construct the singular quadratures when the singulariy is at a point inside the element.  

### Reference line

### Reference triangle

Given a point $x$ inside of $\hat{\tau}$, we simply decompose it into three triangles $\tau_i$, and construct for each sub-triangle the affine map from the reference element $\hat{\tau} \to \tau_i$, being careful to place the point $x$ at the vertex $(1,0)$ of the reference triangle. Here is what this looks like in practice:

```julia
tri = ReferenceTriangle()
x   = (0.3,0.2)
tris = decompose(tri,x)
plot(tris[1],lc=:red,label="",aspect_ratio=:equal)
plot!(tris[2],lc=:blue,ls=:dot,label="")
plot!(tris[3],lc=:green,ls=:dash,label="")
scatter!(x,label="singular point",m=:o)
```

Developing a quadrature for each simplex, and mapping them back to the reference triangle, we obtain the following:

```julia
qreg  = TensorProductQuadrature(GaussLegendre(5),GaussLegendre(5))
shand = Duffy()
quad = SingularQuadratureRule(qreg,shand)
x1,w1 = quad(tris[1])
x2,w2 = quad(tris[2])
x3,w3 = quad(tris[3])
scatter!(x1,label="triangle 1",m=:x)
scatter!(x2,label="triangle 2",m=:x)
scatter!(x3,label="triangle 3",m=:x)
```

### Reference square

For the reference square, note that we can decompose it as four triangles, or four squares. In each case we are led back to the base cases we already considered:

```julia
sq  = ReferenceSquare()
x̂   = (0.3,0.7)
tris = decompose(sq,x̂)
fig_tri = plot(aspect_ratio=:equal)
plot!(fig_tri,x̂,label="",m=:o)    
for tri in tris
    xq,_ = quad(tri)
    plot!(fig_tri,tri,label=:none)
    scatter!(xq,label=:none,m=:x)
end
plot(fig_tri)
```

```julia
shand_1d = Kress(order=2)
shand    = TensorProductSingularityHandler(shand_1d,shand_1d)
q1d      = GaussLegendre(5)
qreg     = TensorProductQuadrature(q1d,q1d)
quad     = SingularQuadratureRule(qreg,shand)

sq  = ReferenceSquare()
sqs = decompose(sq,x̂,sq)
fig_sq = plot(aspect_ratio=:equal)
plot!(fig,x̂,label="singular point",m=:o)    
for sq in sqs
    xq,_ = quad(sq)
    plot!(fig_sq,sq,label=:none)
    scatter!(xq,label=:none,m=:x)
end
plot(fig_tri,fig_sq)
```

## Physical element

So far we have considered how to perform the singular integration over the reference element with a singular point inside by decomposing it into the base case where the singular point is on a predetermined vertex. We will now illustrate how this can be used in conjunction with the map $\chi_\tau : \hat{\tau} \to \tau$ to integrate over an actual physical element in two- or three-spatial dimensions. For that, let us create a sphere with a mesh composed of linear elements:

### Triangular elements

In this first part, we will consider triangular elements in three-dimensions. We will use `gmsh` as our default mesher. We will look at linear elements first simply because this will make our visualization easier. 

```julia
gr()
Geometry.clear!()
Ω,M = WaveProp.IO.gmsh_sphere(dim=2,h=0.2,order=1);
Γ = boundary(Ω)
plot(view(M,Γ))
```

We will extract one triangular element of our mesh. Under the hood, our mesh is composed of three types of elements: 
- Point
- Line
- Triangles

```julia
EE = etypes(M)
```

To extract a triangle we will simply create an iterator over all the elements of type, and extract e.g. the first triangle

```julia
els = ElementIterator(M,EE[2]);
τ   = els[1]
plot(τ,lc=:black)
plot(τ)
```

As before, we will now generate a quadrature over the reference element assuming a singular point $\hat{x}$ inside, and map that quadrature to the surface using the parametrization to see what looks like:

```julia
# genenerate a singular quadrature rule on the base cas
qreg  = TensorProductQuadrature(GaussLegendre(5),GaussLegendre(5))
shand = Duffy()
quad = SingularQuadratureRule(qreg,shand);

# decompose the element's domain into base cases
τ   = els[1]
τ̂   = domain(τ)
x̂   = (0.5,0.2) # singular point on reference domain
tris = decompose(tri,x̂)

# compute the nodes on the reference element
fig_ref = plot(aspect_ratio=:equal,title="τ̂")

plot!(fig_ref,tris[1],lc=:red,label="")
plot!(fig_ref,tris[2],lc=:blue,ls=:dot,label="")
plot!(fig_ref,tris[3],lc=:green,ls=:dash,label="")
scatter!(fig_ref,x̂,label="singular point",m=:o)

x̂1,w1 = quad(tris[1])
x̂2,w2 = quad(tris[2])
x̂3,w3 = quad(tris[3])
scatter!(fig_ref,x̂1,label="",m=:x)
scatter!(fig_ref,x̂2,label="",m=:x)
scatter!(fig_ref,x̂3,label="",m=:x)

# map the nodes to the physical element
x   = τ(x̂)
fig_phys = plot(τ,lc=:black,title="τ")
scatter!(fig_phys,τ.(x̂1),label="",m=:x)
scatter!(fig_phys,τ.(x̂2),label="",m=:x)
scatter!(fig_phys,τ.(x̂3),label="",m=:x)
scatter!(fig_phys,[x],label="singular point",m=:o)
plot(fig_ref,fig_phys)
```

### Quadrialeteral elements

```julia
pyplot()
geo = Geometry.Sphere()
M   = meshgen(Domain(geo))
EE = etypes(M)
els = ElementIterator(M,EE[1]);
plot(geo)
```

```julia
pyplot()

τ        = els[1]
τ̂        = domain(τ)

x̂   = (0.3,0.4) # singular point on reference domain
x   = τ(x̂)

sqs = decompose(τ̂,x̂,τ̂)

# the singular quadrature rule
shand_1d = Kress(order=2)
shand    = TensorProductSingularityHandler(shand_1d,shand_1d)
q1d      = GaussLegendre(10)
qreg     = TensorProductQuadrature(q1d,q1d)
quad     = SingularQuadratureRule(qreg,shand)

fig_ref = plot(aspect_ratio=:equal,title="τ̂")
for sq in sqs
    x̂q,_ = quad(sq)
    plot!(fig_ref,sq,label=:none)
    scatter!(fig_ref,x̂q,label=:none,m=:x)
end

fig_phys = plot(τ,color=:lightgray;h=0.05,camera=(250,170),title="τ")
#fig_phys = plot(title="τ")
for sq in sqs
    x̂q,_ = quad(sq)
    xq = τ.(x̂q)
    scatter!(fig_phys,xq,label=:none,m=:o,ms=:2)
end
scatter!([x],mc=:red,m=:x)
plot(fig_ref,fig_phys) 
```
